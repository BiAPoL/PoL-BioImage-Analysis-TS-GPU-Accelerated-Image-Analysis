```python
import cupy as cp
import numpy as np
from skimage.io import imread
import stackview
```

```python
image = imread("../../data/blobs.tif")
```

```python
image1 = cp.asarray(image[1:])
image2 = cp.asarray(image[:-1])
```

```python
squared_difference = cp.ElementwiseKernel(
    'T x, T y',
    'T z',
    'z = (x - y) * (x - y)',
    'squared_difference')
```

```python
sqdiff = squared_difference(image1, image2)
```

```python
stackview.insight(sqdiff)
```

```python
stackview.picker(sqdiff)
```
