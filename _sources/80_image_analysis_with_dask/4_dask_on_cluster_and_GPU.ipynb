## Practical 4: Dask on the cluster

```python
from distributed import LocalCluster, Client

lc = LocalCluster(n_workers=1, threads_per_worker=4, resources={'GPU': 1})
client = Client(lc)

client
```

```python
import dask
import dask.array as da

N = 1000
c = 100
```

```python
%%timeit -r 1

with dask.annotate(resources={'GPU': 1}):
    y = da.zeros([N] * 2, chunks=c)
    y.compute()



# y.compute(optimize_graph=False)
```
